```python
# Important Analysis Questions
```

```python
#Which country has highest total cases 

letest=df[df['date']==df['date'].max()]
print(letest.sort_values('total_cases', ascending=False))
```

```python
#India max new cases day
print(india.loc[india['new_cases'].idxmax()])
```

```python
#Total deaths per country

print(df.groupby('location')['total_deaths'].max())
```

```python
#Death rate

df['death_rate']=df['total_deaths']/df['total_cases']
print(df[['location','date','death_rate']].tail())

```

```python
# Cases VS Population

df['infected_percent']=df['total_cases']/df['population']*100
df['infected_percent']
```

```python
plt.Figure(figsize=(10,5))
plt.plot(india['date'],india['new_cases'])
plt.title('india daily cases')
plt.xlabel('Date')
plt.ylabel('Cases')
plt.show()
```

```python
for county in df['location'].unique():
    temp=df[df['location']==county]
    plt.plot(temp['date'],temp['total_cases'], label=county)
plt.legend()
plt.show()
```

```python
df['date']=pd.to_datetime(df['date'])
latest=df[df['date']==df['date'].max()]
plt.Figure(figsize=(7,7))
plt.pie(latest['total_cases'],labels=latest['location'],autopct='%1.1f%%')
plt.title('Total Covid cases Share')
plt.show()
```

```python
plt.pie(latest['total_deaths'],labels=latest['location'],autopct='%1.1f%%')
plt.title('Death share by country')
plt.show()
```

```python
df['date'] = pd.to_datetime(df['date'])
latest=df[df['date']==df['date'].max()]
plt.Figure(figsize=(8,5))
plt.bar(latest['location'],latest['total_cases'])
plt.title('Total Covid cases Share')
plt.xlabel('country')
plt.ylabel('cases')

plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```
